```python
import pulp
import math

prob1 = pulp.LpProblem('Problema 1', pulp.LpMinimize)

x1 = pulp.LpVariable('Posto 1', lowBound=0, cat='Binary')
x2 = pulp.LpVariable('Posto 2', lowBound=0, cat='Binary')
x3 = pulp.LpVariable('Posto 3', lowBound=0, cat='Binary')
x4 = pulp.LpVariable('Posto 4', lowBound=0, cat='Binary')
x5 = pulp.LpVariable('Posto 5', lowBound=0, cat='Binary')

custo = x1 + x2 + x3 + x4 + x5

prob1 += custo

sul = x1 + x2

prob1 += (sul>=1)

central = x1 + x2 + x3 + x4

prob1 += (central>=1)

sudeste = x2 + x5

prob1 += (sudeste>=1)

oeste = x3 + x4 + x5

prob1 += (oeste>=1)

norte = x4 + x5

prob1 += (norte>=1)


print (prob1)

optimization_result = prob1.solve()

assert optimization_result == pulp.LpStatusOptimal

for var in (x1,x2,x3,x4,x5):
    print(' A quantidade ótima de {} é: {:1.3f}'.format(var.name, var.value()))
```
